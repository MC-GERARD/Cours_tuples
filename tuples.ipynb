````Python
def division(a,b):
    """Cette fonction donne le quotient et le reste de la division  eucliedienne de a par b, a et b doivent être positif"""
    r=a
    q=0
    while r>=b:
        r = r-b
        q = q + 1
    return q,r
x,y = division(25,7)
print(x)
print(y)
````
